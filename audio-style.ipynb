```python
import tensorflow as tf
import librosa
import os
from IPython.display import Audio, display
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N_FFT = 2048
def read_audio_spectrum(filename):
    x, fs = librosa.load(filename)
    S = librosa.stft(x, N_FFT)
    p = np.angle(S)
    
    S = np.log1p(np.abs(S[:,:430]))  
    return S, fs
```

```python
audio_file = "./database/data/balanced_train/_-_6VO7DIFU_20000_30000.flac"
audio_file = "./database/data/balanced_train/_jOC7pLwiDc_30000_40000.flac"
audio_file = "./database/data/balanced_train/6U7ZxVxc-a0_30000_40000.flac"
a, fs = read_audio_spectrum(audio_file)

N_SAMPLES = a.shape[1]
N_CHANNELS = a.shape[0]
```
